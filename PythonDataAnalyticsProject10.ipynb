# 22. Find Unique Values From Genre 


```python
df["Genre"].unique()
```

```python
list1 = []

for value in df["Genre"]:
    list1.append(value.split(","))
```

```python
print(list1)
```

```python
one_dimension = []

for item in list1:
    for item1 in item:
        one_dimension.append(item1)
```

```python
print(one_dimension)
```

```python
uni_list = []

for item in one_dimension:
    if item not in uni_list:
        uni_list.append(item)
```

```python
uni_list
```

```python
len(uni_list)
```
