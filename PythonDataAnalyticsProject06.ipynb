# 12. Bivariate Analysis


## How Has Better Chance of Survival Male or Female?


```python
df.head()
```

```python
survival_based_on_sex = df.groupby("Sex")["Survived"].sum()
```

```python
survival_based_on_sex.plot(kind="bar")
```

```python
sns.barplot(data=df, x="Sex", y="Survived")
```

## Which Passenger Class Has Better Chance of Survival (First, Second, Or Third Class)? 


```python
df.head()
```

```python
survival_based_on_class = df.groupby("Pclass")["Survived"].sum()
```

```python
survival_based_on_class.plot(kind="bar")
```

```python
df.groupby("Pclass")["Survived"].sum()
```
