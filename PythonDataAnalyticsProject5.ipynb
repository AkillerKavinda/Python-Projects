# 17. Replace Salary Values With 0 and 1


```python
df.head()
```

```python
df.dtypes
```

```python
df.replace?
```

```python
help(df.replace)
```

```python
def salary_values (income):
    if income=="<=50K":
        return 0
    else:
        return 1

df["Salary Values"] = df["income"].apply(salary_values)
```

```python
df.head()
```

```python
# Other way - df["Salary Values"] = df["income"].apply(lambda x: 0 if x == "<=50K" else 1)
```

```python
sns.countplot(y="gender", hue="Salary Values", data=df)
```
