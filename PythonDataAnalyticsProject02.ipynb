# Practicing Joins

```python
df1 = pd.DataFrame({
    "CustomerID": [1, 2, 3, 4],
    "Name": ["Alice", "Bob", "Charlie", "David"],
    "Country": ["USA", "Canada", "USA", "UK"]
})

# Second DataFrame for joining
df2 = pd.DataFrame({
    "CustomerID": [3, 4, 5, 6],
    "TotalSales": [250, 400, 150, 300],
    "Membership": ["Gold", "Silver", "Bronze", "Gold"]
})
```

```python
df_inner = pd.merge(df1, df2, on="CustomerID", how="inner")
```

```python
df_inner
```

```python
df_outer = pd.merge(df1, df2, on="CustomerID", how="outer")
```

```python
df_outer
```

```python
df_left = pd.merge(df1, df2, on="CustomerID", how="left")
```

```python
df_left
```

```python
df_right = pd.merge(df1, df2, on="CustomerID", how="right")
```

```python
df_right
```

# Pivot Tables

```python
df = pd.DataFrame({
    "Region": ["East", "East", "West", "West", "East"],
    "Product": ["A", "B", "A", "B", "A"],
    "Sales": [100, 150, 200, 250, 300],
    "Quantity": [10, 15, 20, 25, 30]
})
```

```python
pd.pivot_table(df, index="Region", values="Sales",aggfunc="sum")
```

```python
pd.pivot_table(df, index="Product", values="Quantity", aggfunc="max")
```

```python
pd.pivot_table(df, index="Product", values="Quantity", aggfunc="min")
```

```python
pd.pivot_table(df, index="Product", values="Quantity", aggfunc="mean")
```

```python
pd.pivot_table(df, index="Product", values="Quantity", aggfunc="median")
```

```python
df.head()
```

```python
pd.pivot_table(df, index=["Region", "Product"], values="Sales", aggfunc="sum")
```

```python
df
```

```python
import matplotlib.pyplot as plt
```

```python
regional_sales = df.groupby("Region")["Sales"].sum()
plt.figure(figsize=(5,5))
regional_sales.plot(kind="bar", color="skyblue")
plt.title("Sales Per Region")
plt.xlabel("Sales")
plt.ylabel("Region")
plt.xticks(rotation=75)
plt.tight_layout()
plt.show()
```

```python
df.head()
```

```python
sales_by_quantity = pd.pivot_table(df, index=["Region", "Product"], values=["Quantity"], aggfunc= "sum")
```

```python
sales_by_quantity.plot(kind='bar',color="purple")

plt.title("Sales by Region and Product")
plt.ylabel("Total Sales")
plt.xlabel("Region and Product")
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
