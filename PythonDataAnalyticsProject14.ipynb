# Data Analysis

## Univariate Analysis - Data distributions of each of the columns

```python
sns.boxplot(data=df, y="price")
plt.show()
```

```python
# Making a new dataframe without the outlier

df = df[df["price"]<1500]
df
```

```python
sns.boxplot(data=df, y='price')
```

```python
# Price distribution

plt.figure(figsize=(8,5))
sns.histplot(data=df, x='price', bins=100)
plt.title("Price Distribution")
plt.ylabel("Frequency")
plt.show()
```

```python
df.columns
```

```python
# Availability distribution

plt.figure(figsize=(6,3))
sns.histplot(data=df, x="availability_365")
plt.title("Availability Distribution")
plt.ylabel("availability_365")
plt.show()
```

```python
df.dtypes
```

### Feature engineering

```python
# Finding the mean prices in the neighbourhood group

df.groupby(by="neighbourhood_group")["price"].mean()
```

```python
df["price_per_bed"] = df["price"]/df["beds"]
df
```

```python
# Average price per bed

df.groupby(by="neighbourhood_group")["price_per_bed"].mean()
```

```python
# Heatmap - correlation of one variable with others based on numerical columns

corr = df[['latitude', 'longitude', 'price', 'minimum_nights', 'reviews_per_month', 'availability_365','beds']].corr()
corr
```

```python
plt.figure(figsize=(8,6))
sns.heatmap(data=corr, annot=True)
plt.show()
```
