```python
# Best Selling authors

df["Total Revenue"] = df["Price"] * df["Review Count"]
author_sales = df.groupby("Author")["Total Revenue"].sum().head(10)

top_10_authors = author_sales.sort_values(ascending=False).head(10)
```

```python
sns.barplot(data=top_10_authors)
plt.xticks(rotation=75)
plt.show()
```

```python
# Finding the author with maximum reviews

df.groupby("Author")["Review Count"].sum().sort_values(ascending=False).head(1)
```

```python
df.groupby("Author")["Review Count"].sum().sort_values(ascending=False).head(1).index[0]
```

```python
author_reviews = df.groupby('Author')['Review Count'].sum()

max_reviews_author = author_reviews.idxmax() # Gives the name
max_reviews_value = author_reviews.max() # Gives the number

print(f"Author with max reviews: {max_reviews_author} with {max_reviews_value} reviews ")
```

```python
# Plotting the authors with max reviews

top_10_authors_with_max_reviews = author_reviews.sort_values(ascending=False).head(10)

sns.barplot(data=top_10_authors_with_max_reviews)
plt.title("Top 10 Authors with Most Reviews")
plt.xticks(rotation=75)
plt.show()
```
