# 20. Plot Continuous Variables


```python
cate_val = []
cont_val = []

for column in df.columns:
    if df[column].nunique() <= 10:
        cate_val.append(column)
    else:
        cont_val.append(column)
```

```python
cate_val
```

```python
cont_val
```

```python
df.hist(cont_val, figsize=(10,10))
plt.tight_layout()
plt.show()
```
